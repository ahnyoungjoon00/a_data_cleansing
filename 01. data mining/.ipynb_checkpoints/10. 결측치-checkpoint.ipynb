## 데이터 제외하기 dropna

```python
import pandas as pd
df = pd.read_excel("score.xlsx", index_col = "지원번호")
df
```

```python
df.dropna() # NaN을 포함하는 행 전체를 삭제
```

```python
df
```

```python
df.dropna(inplace = True) # NaN을 포함하는 행 전체를 삭제
```

```python
df
```

```python
df = pd.read_excel("score.xlsx", index_col = "지원번호")
df
```

#### axis = index or columns
#### how : any or all

```python
df.dropna(axis = "index", how = "any") # NaN 잇는 데이터의 행을 삭제할지 열을 삭제할지
```

```python
df.dropna(axis = "columns") 
```

```python
df["학교"] = np.nan
df
```

```python
df.dropna(axis = "columns", how = "all") # how = "all"의 경우는 데이터 전체가 NaN인 경우에만 삭제하는거
```
