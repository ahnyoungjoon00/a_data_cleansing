# 11. 데이터 정렬

```python
import pandas as pd
df = pd.read_excel("score.xlsx", index_col = "지원번호")
df
```

```python
df.sort_values("키")
```

```python
df.sort_values("키", ascending = False)
```

```python
df.sort_values(["수학", "영어"]) # 수학이 우선순위를 가지고 동점일 경우 영어가 높은게 높은 점수임
```

```python
df.sort_values(["수학", "영어"], ascending = False)
```

```python
df.sort_values(["수학", "영어"], ascending = [True, False]) #수학점수는 오름차순, 영어점수는 내림차순
```

```python
df["키"].sort_values()
```

```python
df["키"].sort_values(ascending = False)
```

```python
df.sort_index() # 인덱스 순서대로 재정렬
```

```python
df.sort_index(ascending = False) # 인덱스 역순서대로 재정렬
```
