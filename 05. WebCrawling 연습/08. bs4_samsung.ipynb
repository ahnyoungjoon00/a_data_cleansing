```python
import requests
from bs4 import BeautifulSoup
import urllib3
urllib3.disable_warnings()
```

```python
url = "https://finance.naver.com/item/sise.naver?code=005930"
```

```python
res = requests.get(url, verify = False)
res.raise_for_status()
```

```python
soup = BeautifulSoup(res.text, "lxml")
```

```python
stocks = soup.find_all("td", attrs = {"class": "num"})
```

```python
price = stocks[4].span.get_text()
print(price)
```

```python
link = stocks[1].span["class"]
print(link)
```

```python
print("https://finance.naver.com/" + str(link))
```

```python
for stock in stocks :
    price = stocks[1].span.get_text()
    link = "https://finance.naver.com/" + str(stocks)
    print(price, link)
```
