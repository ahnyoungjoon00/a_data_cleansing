```python
# 1단계: 사이트 페이지 오픈
# url 로 모든 제어가 가능할때 사용하는 함수

#검색어만 있을 경우 사용
def OpenNewsUrl(search,sdate,edate):   # 넘겨주는 파라메터가 1개이므로 받는 것도 1개이어야 함
#검색어 및 검색기간 입력시 사용
#def OpenNewsUrl(search,sdate,edate):# 넘겨주는 파라메터가 3개이므로 받는 것도 3개이어야 함
    url = "https://news.sbs.co.kr/news/search/main.do?pageIdx=1&searchTermStartDate={sdate}&searchTermEndDate={edate}&searchSection=01%7C02%7C03%7C07%7C08%7C09%7C14&searchCategory=&searchMode=&query={search}&searchOption=4&searchStartDate={sdate}&searchEndDate={edate}"
#검색어만 있을 경우 사용    
    url = url.format(search=search,sdate=sdate,edate=edate)
#검색어 및 검색기간 입력시 사용
#     url = url.format(search=search,sdate=sdate,edate=edate)
    print(url)   # url 확인용
    
    driver = webdriver.Chrome("./chromedriver.exe")
    driver.get(url)
    time.sleep(2)
    
    ListNewsUrl(driver)
```

```python
# 2단계: 목록에서 페이지 이동하면서 url 가져오기

def ListNewsUrl(driver):
    url_temp = driver.current_url   # 현재 브라우저의 url 정보를 가져오는 명령어
    url_temp = url_temp.replace("pageIdx=1","")   # url 에 page가 포함되어 있기에 이것을 제거
    for x in range(1,(15*2)+1,15):
        url = url_temp + "&pageIdx={pageNo}".format(pageNo=x)   # 여기 변수명과 위의 변수명은 같아야 함
        print("리스트: ",url)    # 목록페이지의 url 보는 용도
        driver.get(url)
        time.sleep(2)            # 2초 대기
        
        arrStrUrls =[]
        arrUrls = driver.find_elements_by_css_selector(".psearch_result_list li .psearch_list_w .psil_inner a ")   # 목록페이지에서 url list 가져오기 
        
#         print(len(arrUrls))
#         print(arrUrls[0])
        
        for i in arrUrls:
#             print(i.text)
#             print(i.get_attribute("href"))
            arrStrUrls.append(i.get_attribute("href"))
        
        print("URL List Count:",len(arrStrUrls))       # url list가 몇개인지 확인
        for i in arrStrUrls:                           # 배열의 url list 내용 보기 
            print(i)
        
        NewsInfo(driver,arrStrUrls)
```

```python
def connectDB(dbData):
    DB_HOST = '127.0.0.1'
    DB_USER = 'root'
    DB_PASSWD = 'autoset'
    DB_NAME = 'news'
    
    conn = pymysql.connect(host=DB_HOST, user=DB_USER, password=DB_PASSWD,
                       db=DB_NAME, charset='utf8')
    curs = conn.cursor()
    sql = """insert into yhnews(url,newsDate,newsTitle,newsSubtitle,contents) values (%s, %s, %s, %s, %s)"""
    curs.executemany(sql,dbData)
    conn.commit()

    conn.close()
```
