```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "springgreen", marker = "^", markersize = 9)
plt.xticks([0, 1, 2]) # x축 눈금표시하기
plt.yticks(np.arange(1, 6)) # y축 눈금표시하기
plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "springgreen", marker = "^", markersize = 9)
plt.xticks(np.arange(0, 2, 0.2), labels = ["Jan", " ","Fed", " ", "Mar", "", "May", " ", "June", " ", "July"])
plt.yticks(np.arange(0, 7), ("0", "1GB", "2GB", "3GB", "4GB", "5GB", "6GB")) 
# 그래프의 눈금을 문자열 형태로 지정해서 나타낼 수 있음
plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "springgreen", marker = "^", markersize = 9)
plt.xticks(np.arange(0, 2, 0.2), labels = ["Jan", " ","Fed", " ", "Mar", "", "May", " ", "June", " ", "July"])
plt.yticks(np.arange(0, 7), ("0", "1GB", "2GB", "3GB", "4GB", "5GB", "6GB")) 

plt.tick_params(axis = "x", direction = "in", length = 3, pad = 6, labelsize = 14, labelcolor = "green", top = True)
plt.tick_params(axis = "y", direction = "inout", length = 10, pad = 15, labelsize = 12, width = 2, color = "r")
# 눈금 방향, 길이, 레이블과의 거리, 레이블 크기, 레이블 색상, 눈금 표시위치, 너비, 색상 지정 가능함
plt.show()
```
