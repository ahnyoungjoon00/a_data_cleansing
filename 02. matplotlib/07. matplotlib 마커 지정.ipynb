```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4], [2 ,3, 5, 10], "bo") # point 찍기
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# plt.plot([1, 2, 3, 4], [2, 3, 5, 10], "bo-") # o 점과 - 실선을 동시에
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], "bo--") # o 점과 -- 점선을 동시에
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
plt.plot([4, 5, 6], marker = "H")
plt.plot([3, 4, 5], marker = "d")
plt.plot([2, 3, 4], marker = "x")
plt.plot([1, 2, 3], marker = 11)
plt.plot([0, 1, 2], marker = "$Z$")
# marker 파라미터 사용하기
plt.show()
```
