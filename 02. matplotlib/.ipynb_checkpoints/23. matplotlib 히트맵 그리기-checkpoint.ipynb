```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.show()
```

```python
arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar() # 히트맵의 컬러바를 같이 보여준다

plt.show()
```

```python
arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar(shrink = 0.8, aspect = 10) 
# shrink는 컬러바의 크키 - 디폴트 1.0, aspect는 종횡비 - 디폴트 20
plt.show()
```

```python
arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar(shrink = 0.8, aspect = 10)
plt.clim(-1.0, 1.0)
# plt.clim(-3.0, 3.0)
# 색상의 범위를 설정
plt.show()
```

```python
arr = np.random.standard_normal((30, 40))

# cmap = plt.get_cmap("PiYG")
# cmap = plt.get_cmap("BuGn")
# cmap = plt.get_cmap("Greys")
cmap = plt.get_cmap("bwr")
# 컬러맵 종류를 지정해서 그려주기
plt.matshow(arr, cmap = cmap)
plt.colorbar()
plt.show()
```
