```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10 ,100)
y = x ** 3

plt.plot(x, y)
plt.xscale("symlog") # x축 스케일을 symlog(symmetric log scale)로 지정
plt.show()
```

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
# plt.yscale("linear") # 선형 스케일
plt.yscale("log")  #로그 스케일은 지수함수와 같은 기하급수적으로 변화하는 데이터에 사용용이
plt.show()
```
