```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([2, 3, 5, 10]) # x값은 자동으로 0, 1, 2, 3...로 입력됨
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.show()
```

```python
data_dict = {"data_x" : [1, 2, 3, 4, 5],
            "data_y" : [2, 3, 5, 10, 8]}
plt.plot("data_x", "data_y", data = data_dict) # 파이썬 딕셔너리와 같이 레이블이 있는 경우 레이블을 지정
plt.show()
```
