```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 01.Matplotlib 기본사용

```python
plt.plot([1, 2, 3, 4, 1])
plt.show()
```

```python
# x, y축 data값 설정
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
# x, y축 데이터로 Points 찍기
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') #'ro'는 색상 : red, 형태 :o 
plt.axis([0, 6, 0, 20]) # x, y축 각각의 범위를 설정
plt.show()
```

```python
t = np.arange(0., 5., 0.2) # x축 시작점 0., 200ms 간격으로 균일하게 샘플된 시간

plt.plot(t, t, "r--", t, t**2, "bs", t, t**3, "g^") # 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.show()
```
