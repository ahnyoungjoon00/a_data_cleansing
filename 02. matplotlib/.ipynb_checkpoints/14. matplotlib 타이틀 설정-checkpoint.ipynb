```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "forestgreen", marker = "^", markersize = 9)

plt.tick_params(axis = "both", direction = "in", length = 3, pad = 6, labelsize = 14)
plt.title("Graph Title")
# 그래프 타이틀 붙이기
plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "forestgreen", marker = "^", markersize = 9)

plt.tick_params(axis = "both", direction = "in", length = 3, pad = 6, labelsize = 14)
plt.title("Graph Title", loc = "right", pad = 20)
# 타이틀 위치와 오프셋 지정
plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "forestgreen", marker = "^", markersize = 9)

plt.tick_params(axis = "both", direction = "in", length = 3, pad = 6, labelsize = 14)
plt.title("Graph Title", loc = "right", pad = 20)

title_font = {
    "fontsize" : 16,
    "fontweight" : "bold"
}
plt.title("Graph Title", fontdict = title_font, loc = "right", pad = 20)
# 타이틀 폰트 지정
plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "forestgreen", marker = "^", markersize = 9)

plt.tick_params(axis = "both", direction = "in", length = 3, pad = 6, labelsize = 14)
title_right = plt.title("Graph Title", loc = "right", pad = 20)

title_font = {
    "fontsize" : 16,
    "fontweight" : "bold"
}
title_left = plt.title("Graph Title", fontdict = title_font, loc = "right", pad = 20)

print(title_left.get_position())
print(title_left.get_text())
print(title_right.get_position())
print(title_right.get_text())
# 타이틀 텍스트 위치와 문자열을 반환받는 법
plt.show()
```
