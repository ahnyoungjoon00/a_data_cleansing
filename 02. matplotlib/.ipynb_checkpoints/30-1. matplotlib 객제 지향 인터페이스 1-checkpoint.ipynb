```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
# 위의 plt.subplots()를 다른 방법으로 나타내는 방법
plt.show()
```

```python
fig, ax = plt.subplots(2, 2) # plt.subplots(nrows, ncols)형태로 지정
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, sharex = True, sharey = True) #중복된 축을 한번씩만 표시하도록
plt.show()
```
