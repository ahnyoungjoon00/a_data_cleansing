```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 2, 0.2)
plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "springgreen", marker = "^", markersize = 9)
plt.grid(True) # 그리드(격자)를 형성해줌

plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "forestgreen", marker = "^", markersize = 9)
plt.grid(True, axis = "y")
# 그리드(격자)를 Y축만 그리겠다. (axis = "both/x/y")로
plt.show()
```

```python
x = np.arange(0, 2, 0.2)
plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "springgreen", marker = "^", markersize = 9)
plt.grid(True, axis = "y", color = "red", alpha = 0.5, linestyle = "--")
# 그리드 스타일도 정해줄 수 있음
plt.show()
```
