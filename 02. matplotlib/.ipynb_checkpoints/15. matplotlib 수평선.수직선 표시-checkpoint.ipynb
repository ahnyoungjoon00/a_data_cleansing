```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x+1, "bo")
plt.plot(x, x**2 , "g--")
plt.plot(x, -2*x + 3, "r:")

plt.axhline(4.0, 0.1, 0.9, color = "lightgray", linestyle = "--", linewidth = 2)
plt.hlines(-0.62, 1.0, 2.5, color = "gray", linestyle = "solid", linewidth = 3)
# 선그래프 중간에 수평선 그려주기
plt.show()
```

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x+1, "bo")
plt.plot(x, x**2 -4, "g--")
plt.plot(x, -2*x + 3, "r:")

plt.axvline(1.0, 0.2, 0.8, color = "lightgray", linestyle = "--", linewidth = 2)
plt.vlines(1.8, -3.0, 2.0, color = "gray", linestyle = "solid", linewidth = 3)
# 선그래프 중간에 수직선 그려주기
plt.show()
```
