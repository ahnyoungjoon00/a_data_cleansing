## Matplotlib 축 레이블 설정하기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 10])
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.show() #그림 x, y축의 레이블을 표시해줌
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel("X-Axis", labelpad = 15)
plt.ylabel("Y-Axis", labelpad = 20)
plt.show() # x축 여백 15pt, y축 여백 20pt 지정
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel("X-Axis", labelpad = 15, fontdict = {"family" : "serif", "color" : "blue", "weight" : "bold", "size" : 14})
plt.ylabel("Y-Axis", labelpad = 20, fontdict = {"family" : "fantasy", "color" : "deeppink", "weight" : "normal", "size" : "xx-large"})
plt.show() # 레이블 속성(폰트, 여백, 색상, 굵기, 크기) 결정
```

```python
# 미리 폰트설정 해놓고 간편하게 쓰기

font1 = {"family" : "serif", 
        "color" : "blue", 
        "weight" : "bold", 
        "size" : "14"}
font2 = {"family" : "fantasy", 
        "color" : "deeppink", 
        "weight" : "normal", 
        "size" : "xx-large"
        }
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel("X-Axis", labelpad = 15, fontdict = font1)
plt.ylabel("Y-Axos", labelpad = 20, fontdict = font2)
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel("X-Axis", loc = "right")
plt.ylabel("Y-Axis", loc = "top")
plt.show() # matplotlib 버전이 3.3.2 이상이어야 구현가능 (지금은 굳이 안쓴다)
```
