```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 2, 0.2) # 0부터 2까지 0.2의 간격으로 array 형성

plt.plot(x, x, "r--", x, x**2, "bo", x, x**3, "g-.") # x값 y값 마크 특성 순
plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, "bo")
plt.plot(x, x**2, color = "#e35f62", marker = "*", linewidth = 2)
plt.plot(x, x**3, color = "forestgreen", marker = "^", markersize = 9)
plt.show() # 곡선 스타일 지정
```
