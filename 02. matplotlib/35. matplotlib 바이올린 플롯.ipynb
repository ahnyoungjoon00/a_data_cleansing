```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("default")
plt.rcParams["figure.figsize"] = (4, 3)
plt.rcParams["font.size"] = 12

np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], positions = [2, 3, 4]) # 바이올린플롯
ax.set_ylim(-10.0, 10.0)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xlabel("Data Type")
ax.set_ylabel("Value")

plt.show()
```

```python
# 최대/치소값, 중간값, 평균값 표시

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], showmeans=True) # showmeans 로 값 표시
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
# showmeans 파라미터는 데이터 분포에서 평균값의 위치에 직선을 표시합니다. (Default: False)

# showextrema는 최대/최소값에 직선을 표시합니다. (Default: True)

# showmedians는 중간값에 직선을 표시합니다. (Default: False)
plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

fig, ax = plt.subplots()

violin = ax.violinplot([data_a, data_b, data_c], quantiles=[[0.25, 0.75], [0.1, 0.9], [0.3, 0.7]]) # quantiles 로 분위수를 지정 가능
ax.set_ylim(-10.0, 10.0)
ax.set_xticks(np.arange(1, 4))
ax.set_xticklabels(['A', 'B', 'C'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```
