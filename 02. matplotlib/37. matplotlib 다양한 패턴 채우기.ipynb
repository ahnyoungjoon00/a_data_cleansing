```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("default")
plt.rcParams["figure.figsize"] = (6, 5)
plt.rcParams["font.size"] = 12

x = [1, 2, 3]
y = [1, 2 ,3]

fig, ((ax1, ax2) , (ax3, ax4)) = plt.subplots(2, 2)

ax1.bar(x, y, color = "aquamarine", edgecolor = "black", hatch = "/")
ax2.bar(x, y, color = "salmon", edgecolor = "black", hatch = "\\")
ax3.bar(x, y, color = "navajowhite", edgecolor = "black", hatch = "+")
ax4.bar(x, y, color = "lightskyblue", edgecolor = "black", hatch = "*")
# hatch는 패턴
plt.tight_layout()
plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = [1, 2, 3]
y = [1, 2, 3]

fig, ax = plt.subplots()

bars = ax.bar(x, y, color='lightgray', edgecolor='black')
bars[0].set_hatch('x')
bars[1].set_hatch('O')
bars[2].set_hatch('.')

plt.tight_layout()
plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12

x = [1,2,3]
y = [1,2,3]

for i in range(6):
    hatch_str = "/" * i # hatch_str 을 /, //, ///, ////, ///// 로 패턴 개수 조절 가능

    ax = plt.subplot(2, 3, i + 1)
    ax.set_title("Hatch String: " + hatch_str)
    bars = ax.bar(x,y,facecolor='skyblue', edgecolor='black')

    for bar in bars:
        bar.set_hatch(hatch_str)

plt.tight_layout()
plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12
plt.rcParams['hatch.linewidth'] = 3 # 패턴 선 두께를 조절 가능

x = [1,2,3]
y = [1,2,3]

for i in range(6):
    hatch_str = "/" * i

    ax = plt.subplot(2, 3, i + 1)
    ax.set_title("Hatch String: " + hatch_str)
    bars = ax.bar(x,y,facecolor='skyblue', edgecolor='black')

    for bar in bars:
        bar.set_hatch(hatch_str)

plt.tight_layout()
plt.show()
```
