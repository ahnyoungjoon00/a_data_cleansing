```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi *x2)

plt.subplot(2, 1, 1) # nrow = 2, ncols = 1, index = 1
plt.plot(x1, y1, "o-")
plt.title("1st Graph")
plt.xlabel("time (s)")
plt.ylabel("Damped oscillation")

plt.subplot(2, 1, 2) # nrow = 2, ncols = 1, index = 2
plt.plot(x2, y2, ".-")
plt.title("2nd Graph")
plt.xlabel("time (s)")
plt.ylabel("Undamped")

plt.tight_layout()
plt.show()
```
