## Matplotlib 축 범위 지정하기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xlim([0, 5]) # x축 범위 
plt.ylim([0, 20]) # y축 범위
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.axis([0, 5, 0, 20]) # x, y축의 범위
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
# plt.axis("square")
plt.axis("scaled") # on/off/equal/scaled/tight/auto/normal/image/square
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)

axis_range = plt.axis("scaled") # 각 축의 범위를 표시해줌
print(axis_range) 
plt.show()
```
