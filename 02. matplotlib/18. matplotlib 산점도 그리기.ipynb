```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
# np random 모듈로 0~1 범위의 난수 50개 생성
# scatter()함수로 위치 마커 표시
plt.scatter(x, y)
plt.show()
```

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2 
# 마커 사이즈 지정인데 plot() 에서 markersize = 20 으로 지정하는 것과 scatter 에서 s = 20**2로 지정하는 것은 같다
colors = np.random.rand(n)

plt.scatter(x, y, s = area, c = colors) # scatter() 함수의 s, c는 각각 마커 크기와 색상을 지정함
plt.show()
```

```python
plt.plot([1], [1], "o", markersize = 20, c = "#FF5733")
plt.scatter([2], [1], s = 20**2, c = "#33FFCE")

plt.text(0.5, 1.05, "plot(markersize = 20)", fontdict = {"size" : 14})
plt.text(1.6, 1.05, "scatter(s = 20**2)", fontdict = {"size" : 14})
plt.axis([0.4, 2.6, 0.8, 1.2])
plt.show()
```

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, s = area, c = colors, alpha = 0.5, cmap = "Spectral")
# alpha는 마커의 투명도 조절 / cmap은 피라미터의 컬러맵에 해당하는 문자열 지정 
plt.colorbar()
# 우측에 컬러의 색에 따른 지수를 표현
plt.show()
```
