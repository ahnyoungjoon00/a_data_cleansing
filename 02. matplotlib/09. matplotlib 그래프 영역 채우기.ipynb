```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]

plt.plot(x, y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.fill_between(x[1:3], y[1:3], alpha = 0.3)
# x, y의 범위 / alpha 는 투명도
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 3, 5, 10]
plt.plot(x, y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.fill_betweenx(y[2:4], x[2:4], alpha = 0.3)
# fill_betweenx 는 y축과 그래프 사이를 채워줌
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color = "lightgray", alpha = 0.5)
# 두 그래프 사이 영역 채우기
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.fill([1.9, 1.9, 3.1, 3.1], [1.0, 4.0, 6.0, 3.0], color = "lightgray", alpha = 0.3)
# 다각형 영역을 자유롭게 지정하여 채울 수 있음
plt.show()
```
