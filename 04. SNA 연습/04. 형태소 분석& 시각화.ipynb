```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from konlpy.tag import Kkma
import pickle
```

```python
kkma = Kkma()
```

```python
file = open("data.pickle", mode = "rb")
crawling_data = pickle.load(file)
crawling_data
len(crawling_data)
```

```python
ex_nouns = []
```

```python
for sent in crawling_data :
    for noun in kkma.nouns(sent) :
        ex_nouns.append(noun)
```

```python
ex_nouns
len(ex_nouns)
```

```python
from re import match
```

```python
nouns_count = {}

for noun in ex_nouns :
    if len(str(noun)) > 1 and not(match('^[0-9]',noun)) :
        nouns_count[noun] = nouns_count.get(noun, 0) + 1
nouns_count
```

```python
from collections import Counter

counter = Counter(nouns_count)
top10 = counter.most_common(n = 10)
top10
```
