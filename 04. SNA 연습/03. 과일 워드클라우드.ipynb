```python
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("fruit_vegetable.csv", encoding = "utf-8")
wc = df.set_index("title").to_dict()["count"]
```

```python
wordCloud = WordCloud(font_path = "malgun",
                     width = 400,
                     height = 400,
                     max_font_size = 100,
                     background_color = "white"
                     ).generate_from_frequencies(wc)
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(wordCloud)
plt.axis("off")
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(wordCloud)
plt.axis("off")
```

```python
white_wordcloud = WordCloud(font_path = "malgun",
                           width = 480, height = 480,
                           background_color = "white").generate_from_frequencies(wc)
black_wordcloud = WordCloud(font_path = "malgun",
                           width = 480, height = 480,
                           background_color = "black").generate_from_frequencies(wc)

fig, ax = plt.subplots(1, 2, figsize = (10, 5))
fig.tight_layout()
ax[0].imshow(white_wordcloud)
ax[1].imshow(black_wordcloud)
```
