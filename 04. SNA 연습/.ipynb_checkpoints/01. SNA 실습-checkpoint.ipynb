## 2. 어휘 동시 출현 빈도의 계수화

```python
from nltk import ConditionalFreqDist
```

```python
sentences = ["I love data science and deep learning", "I love science", "I know this code"]
tokens = [word_tokenize(x) for x in sentences]
bgrams = [bigrams(x) for x in tokens]

token = []
for i in bgrams :
    token += ([x for x in i])
    
cfd = ConditionalFreqDist(token)
cfd.conditions()
```

```python
print(cfd["I"])
```

```python
print(cfd["I"]["love"])
```

```python
print(cfd["I"].most_common(1))
```

```python
import numpy as np

freq_matrix = []

for i in cfd.keys() :
    temp = []
    for j in cfd.keys() :
        temp.append(cfd[i][j])
    freq_matrix.append(temp)
freq_matrix = np.array(freq_matrix)

print(cfd.keys())
print(freq_matrix)
```

### - 판다스로 보기 좋게 시각화

```python
import pandas as pd

df = pd.DataFrame(freq_matrix, index = cfd.keys(), columns = cfd.keys())
df.style.background_gradient(cmap = "coolwarm")
```

### nltk 의 ConditionalProbDist
어휘 동시 출현 빈도와 출현 확률 측정

```python
from nltk.probability import ConditionalProbDist, MLEProbDist
```

```python
cpd = ConditionalProbDist(cfd, MLEProbDist)
cpd.conditions()
```

## 4. 중심성(Centrality) 지수

### 연결 중심성(Degree Centrality) - 다른 단어와 동시 출현 빈도가 많은 단어
주로 (i와 직접적으로 연결된 노드 수 / 노드 i와 직간접적으로 연결된 노드 수)

```python
nx.degree_centrality(g)
```

### 위세 중심성(Eigenvecter centrality)
연결된 상대 단어의 중요성에 가중치를 둬서 중요한 단어와 많이 연결되어 있다면 중심성이 높아짐

```python
nx.eigenvector_centrality(g, weight = "weight")
```

### 근접 중심성(Closeness Centrality)
다른 단어에 얼마나 가깝게 있는지, 직간접적으로 연결된 모든 노드들 사이의 거리를 축정(모든 노드 수 -1 / 특정 노드 i에서 모든 노드에 이르는 최단 경로 수를 더한 수)

```python
nx.closeness_centrality(g, distance = "weight")
```

### 매개 중심성(Betweenness Centrality)
한 단어가 다른 단어와 연결망을 구축하는데 얼마나 도움을 주는지,
빈도수가 작더라도 단어간 의미부여 역할이 크기때문에 제거하면 의사소통이 어려워짐
노드 간 최단 경로에서 특정 노드가 등장하는 횟수로 측정, 표준화를 위해 최댓값인 (N-1)*(N-2) / 2

```python
nx.betweenness_centrality(g)
```

```python
def get_node_size(node_values) :
    nsize = np.array([v for v in node_values])
    nsize = 1000 * (nsize - min(nsize)) / (max(nsize) - min(nsize))
    
    return nsize
```
